# **Colores y Estilos**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.5,11)
```

```python
# Muestra los estilos que tiene matplot
print(plt.style.available)
```

```python
# Usar un estilo de grilla

plt.style.use('seaborn-whitegrid')
```

```python
fig,ax = plt.subplots(figsize =(8,8))
```

```python
# Personalizar colores y estilos en matlab
fig, ax = plt.subplots(figsize =(6,6))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'bo-')
ax.plot(x,x+3,'g.:')
ax.plot(x,x+4,'purple')
plt.show()
```

```python
# Personalizar colores y estilos en Pyplot
fig, ax = plt.subplots(figsize =(6,6))
ax.plot(x,x+1, color = 'green')
ax.plot(x,x+2,color = '#66FF89') # Color RGB
ax.plot(x,x+3,color = 'blue')
ax.plot(x,x+4,color = 'black')
plt.show()
```

```python
# Transparencia (alpha) y grosor (linewidth)
fig, ax = plt.subplots(figsize =(6,6))
ax.plot(x,x+1, color = 'green', alpha = 0.5, linewidth = 18)
ax.plot(x,x+2,color = '#66FF89', linewidth= 3) # Color RGB
ax.plot(x,x+3,color = 'blue', linewidth = 5)
ax.plot(x,x+4,color = 'black', alpha = 0.3, linewidth = 12)
plt.show()
```

```python
# Estilo (linestyle) y Marcadores (marker)
fig, ax = plt.subplots(figsize =(6,6))
ax.plot(x,x+1, color = 'green', linestyle = 'dashed')
ax.plot(x,x+2,color = '#66FF89', linestyle = '--', marker = 'P') # Color RGB
ax.plot(x,x+3,color = 'blue', linewidth=5, linestyle='--')
ax.plot(x,x+4,color = 'black', linewidth=2, linestyle=':')
plt.show()
```

```python
# Tamaño de marcadores (markersize) y color (makerfacecolor)
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "green")

ax.plot(x,x+4, color = 'black',linestyle = '--', marker = 'P', markerfacecolor="red")
plt.show()
```
