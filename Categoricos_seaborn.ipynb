```python
import seaborn as sns
import matplotlib.pyplot as plt 

tips = sns.load_dataset('tips')
tips.head(2)
```

```python
#Gr√°fica de cantidad
sns.countplot(data = tips, x= 'day', hue='sex'); 
```

```python
#Gr√°fica de stripplot
sns.stripplot(data = tips, x= 'day',y='total_bill', hue='sex',dodge=True); 
```

```python
#Diagrama de puntos
sns.swarmplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex',dodge=True);
```

```python
#Diagrama de caja
sns.boxplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True);
```

```python
#Se pueden combinar distintos gr√°ficos
plt.figure(figsize=(6,6))
sns.boxplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True);
sns.swarmplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True,color= '0')
plt.show()
```

```python
#Viol√≠n: parecido a Boxplot, pero no muestra los cuartiles, sino como se concentran los datos con comparaci√≥n split
sns.violinplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', split=True,palette='pastel');
```

```python
#Catplot, con kind podemos cambiar el tipo de gr√°fico y se puede agregar otra variable col
sns.catplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True,kind="box",col="time");
```

```python
#personas que participaron del encuentro para comer.
sns.catplot(data=tips, x="day", hue="size", col="time", kind= "count" )
plt.show()
```
