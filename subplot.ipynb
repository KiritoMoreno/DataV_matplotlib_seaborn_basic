```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

```python
#Crear una matriz de gráficos de una fila y dos columnas
plt.subplot(1,2,1) # gráfico 1
plt.plot(x,y,'r--')
plt.subplot(1,2,2) # gráfico 2
plt.pie(y)
plt.show
```

```python
#Agregar diferentes Plot al gráfico
plt.subplot(1,2,1)
plt.plot(x,y, 'r--') # Plot 1
plt.plot(y,x, 'b:') #Plot 2

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
#Invertir la matriz de los gráficos
plt.subplot(2,1,1) #Grafico 1
plt.plot(x,y,'r--') #Plot 1
plt.plot(y,x,'b:')  #Plot 2

plt.subplot(2,1,2) #Grafico 2
plt.pie(y)         #Plot 1  
plt.show()
```

```python
# Modificar dimensiones (ancho, largo)
plt.figure(figsize=(8, 5))

plt.subplot(2, 2, 1)
plt.plot(x, y, 'ro-')
plt.plot(y, x, 'bv-')
plt.subplot(1, 2, 2)
plt.hist(y)
plt.subplot(2,2,3)
plt.boxplot(x)

plt.show()
```
