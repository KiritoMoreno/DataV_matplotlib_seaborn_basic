```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
#Creamos nuestras gr√°ficas comunes
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
plt.show()
```

```python
#Para crear los t√≠tulos realiza lo siguiente:

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax1.set_title("Relacion X - Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
```

```python
#Agrega labels en los ejes
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', )
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
```

```python
#Cambiar el tama√±o a las figuras
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (5,5))
```

```python
#Aplicar etiquetas (por defecto pone la mejor opci√≥n)
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', label = "sin(x)")
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
plt.show()
```

```python
#Modificar la localizaci√≥n de la etiqueta
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', label = "sin(x)")
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend(loc = "upper center")

ax2.plot(x,y, 'r', label = "$sin(y)$") #anotacion matematica($)
ax2.set_title("Relacion Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
ax2.legend()
plt.show()
```

```python
#Si no queremos usar el metodo objecto
plt.plot(x,y,label= 'sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', box_ancho=(1.0,0.9))
```
